```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model
%matplotlib inline

#plt.style.use('seaborn-whitegrid')
```

```python
Kotchi = pd.read_excel('autoMPG_data.xlsx')
Kotchi.head()
```

```python
y = Kotchi['GallonsPer100Miles']
y
```

```python
m = Kotchi.cov().iloc[0][1]/Kotchi.Displacement100ci.var()

x = Kotchi.Displacement100ci.mean()
y = Kotchi.GallonsPer100Miles.mean()

x,y,m

b = y - m * x
print(b)
print(m)
```

```python
%matplotlib inline
plt.xlabel('Displacement100ci')
plt.ylabel('GallonsPer100Miles')
plt.xlim(0, 6)
plt.ylim(0, 12)
arr = np.array(Kotchi.Displacement100ci)
m = Kotchi.cov().iloc[0][1]/Kotchi.Displacement100ci.var()

x = Kotchi.Displacement100ci.mean()
y = Kotchi.GallonsPer100Miles.mean()

x,y,m

y =  m * arr + b
y
plt.scatter(Kotchi.Displacement100ci, Kotchi.GallonsPer100Miles, color = 'blue', marker = '.')
plt.plot(arr,y, color = 'red')
```

```python
m = Kotchi.cov().iloc[0][1]/Kotchi.Horsepower100.var()

x = Kotchi.Horsepower100.mean()
y = Kotchi.GallonsPer100Miles.mean()

x,y,m

b = y - m * x
print(b)
print(m)
```

```python
%matplotlib inline
plt.xlabel('Horsepower100')
plt.ylabel('GallonsPer100Miles')
plt.xlim(0, 6)
plt.ylim(0, 12)
arr = np.array(Kotchi.Horsepower100)
m = Kotchi.cov().iloc[0][1]/Kotchi.Horsepower100.var()

x = Kotchi.Horsepower100.mean()
y = Kotchi.GallonsPer100Miles.mean()

x,y,m

y =  m * arr + b
y
plt.scatter(Kotchi.Horsepower100, Kotchi.GallonsPer100Miles, color = 'blue', marker = '.')
plt.plot(arr,y, color = 'red')
```
